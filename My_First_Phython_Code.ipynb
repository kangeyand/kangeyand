```python
# importing pandas as pd
import pandas as pd
```

```python
pd.options.display.float_format = "{:,.2f}".format
pd.options.display.max_columns = None
```

```python
# Creating the dataframe 
df = pd.read_csv('Apr 2021 Target For Upload.csv', index_col=0)
```

```python
# Change data types so that all IDs are treated as strings. Only the genuine numbers should be treated as int/float.
data_types = {
 'QTY': 'float64',
 'Target Value': 'float64',
 'Target Qty 1': 'float64',
 'Target Value 1': 'float64',
 'Target Qty 2': 'float64',
 'Target Value 2': 'float64'   
}
```

```python
# Print the dataframe
df
```

```python
# Aggregate group by Store
#df.groupby('Store').count()
groupby_columns = ['Store']
output_columns = ['QTY', 'Target Value', 'Target Qty 1','Target Value 1','Target Qty 2','Target Value 2']
target_by_store = df.groupby(groupby_columns, dropna=False)[output_columns].sum().reset_index()
target_by_store
```

```python
# Writting summary data to new file and reading summary data
summary = df.groupby('Store').sum()
filename="store_summary.csv"
summary.to_csv(filename, index=False)
pd.read_csv(filename, nrows=25)
```
